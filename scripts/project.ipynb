```python
import docx

def extract_text_from_docx(docx_path):
    text = ""
    doc = docx.Document(docx_path)
    for paragraph in doc.paragraphs:
        text += paragraph.text + '\n'
    return text

# Example usage:
text = extract_text_from_docx('data/resumes/Abiral_Pandey_Fullstack_Java.docx')
print(text)
```

```python
from pdfminer.high_level import extract_text

text = extract_text("data/resumes/Anupama_resume.pdf")
print(text)
```

```python
RESUME_SECTIONS_GRAD = [
                    'accomplishments',
                    'experience',
                    'education',
                    'interests',
                    'projects',
                    'professional experience',
                    'publications',
                    'skills',
                    'certifications',
                    'objective',
                    'career objective',
                    'summary',
                    'leadership'
                ]
def extract_entity_sections_grad(text):
    '''
    Helper function to extract all the raw text from sections of
    resume specifically for graduates and undergraduates

    :param text: Raw text of resume
    :return: dictionary of entities
    '''
    text_split = [i.strip() for i in text.split('\n')]
    # sections_in_resume = [i for i in text_split if i.lower() in sections]
    entities = {}
    key = False
    for phrase in text_split:
        if len(phrase) == 1:
            p_key = phrase
        else:
            p_key = set(phrase.lower().split()) & set(RESUME_SECTIONS_GRAD)
        try:
            p_key = list(p_key)[0]
        except IndexError:
            pass
        if p_key in RESUME_SECTIONS_GRAD:
            if p_key not in entities:
                entities[p_key] = []
            key = p_key
        elif key and phrase.strip():
            entities[key].append(phrase)
    return entities
```

```python
t = extract_entity_sections_grad(text)
```
